## `Analisando a distribuiçao dos dados`

```python
import numpy as np

def calcular_largura_bin(df, coluna):
    Q75, Q25 = np.percentile(df[coluna], [75 ,25])
    IQR = Q75 - Q25
    largura_bin = 2 * IQR * np.power(len(df[coluna]), -1/3)
    return largura_bin
```

```python
largura = calcular_largura_bin(dados, 'arrival_time')
sns.histplot(data=dados, x='arrival_time', kde=True, binwidth=largura)
```

```python
largura = calcular_largura_bin(dados, 'departure_time')
sns.histplot(data=dados, x='departure_time', kde=True, binwidth=largura)
```

```python
mean_delay = dados['delay'].mean()
median_delay = dados['delay'].median()

fig, axes = plt.subplots(1, 2, figsize=(9, 4))

sns.boxplot(data=dados, y='delay', ax=axes[0])
axes[0].set_title('Boxplot')

axes[0].axhline(y=mean_delay, color='r', linestyle='--', label='Média')
axes[0].legend()

largura_bin_delay = calcular_largura_bin(dados, 'delay')
sns.histplot(data=dados, x='delay', ax=axes[1], kde=True, binwidth=largura_bin_delay)
plt.ylabel('Número de voos')
plt.grid(False)
axes[1].set_title('Histograma')

axes[1].axvline(x=mean_delay, color='r', linestyle='--', label='Média')
axes[1].axvline(x=median_delay, color='y', linestyle='--', label='Mediana')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## `Feature Engineering`


```python
dados['date'] = pd.to_datetime(dados['year'].astype(str) + '-' + (dados['day'] + 1).astype(str), format='%Y-%j')
dados.head()
```

```python
dados['is_weekend'] = dados['date'].dt.weekday.isin([5,6])
```

```python
dados['day_name'] = dados['date'].dt.day_name()
dados.head()
```

```python
#numero de valores unicos
#Procurando categoricos binarios
dados.nunique()
```

```python
dados['schengen'].unique()
```

```python
dados['schengen'] = dados['schengen'].replace({'non-schengen': 0, 'schengen': 1})
dados['is_holiday'] = dados['is_holiday'].replace({False: 0, True: 1})
dados['is_weekend'] = dados['is_weekend'].replace({False: 0, True: 1})
dados.head()
```

```python
#lista com o nome das variaveis categoricas
variaveis_categoricas = ['airline', 'aircraft_type', 'origin', 'day_name']
```

```python
#Cria categoricos binarios a partir de variaveis categoricas para auxiliar na construçao do modelo de machine learning
df_encoded = pd.get_dummies(data=dados, columns=variaveis_categoricas, dtype=int)
df_encoded.head()
```

```python
#Correlaçao varia de -1 ate 1.
#Quanto mais proximo de 1 é uma positiva perfeita. Diretamente proporcionais
#Quanto mais proximo de -1 é uma negativa perfeita. Inversamente proporcionais
df_encoded[['arrival_time', 'departure_time']].corr()
#Como possuem um indice de correlaçao alto, pode-se eliminar um deles do conjunto de dados
```

```python
df_clean = df_encoded.drop(['flight_id', 'departure_time', 'day', 'year', 'date'], axis=1)
df_clean.head()
```

```python
#variavel explicativa
#retira da tabela a variavel de estudo que é o delay
X = df_clean.drop(['delay'], axis=1)
#variavel target
#define como variavel a ser prevista o delay
y = df_clean['delay']
```

```python
#importa a biblioteca responsavel por realizar o treinamento do modelo
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
#importe do modelo utilizado. Modelo de regressao.
#O DummyRegressor não tenta aprender a partir dos dados; ao invés disso, faz previsões baseadas em estratégias triviais,
#como prever sempre a média do alvo (target), mediana, entre outras.
from sklearn.dummy import DummyRegressor

model_dummy = DummyRegressor()
#O método fit faz com que o DummyRegressor armazene os valores necessários para fazer previsões baseadas na estratégia escolhida
# (por exemplo, a média dos valores em y_train)
model_dummy.fit(X_train, y_train)
```

```python
#X_test contém as características (features) dos dados de teste que não foram usadas durante o treinamento do modelo.
y_pred_dummy = model_dummy.predict(X_test)
y_pred_dummy
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def calcular_metricas_regressao(y_test, y_pred):

    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    metricas = {
        'Raiz do Erro Quadrático Médio': round(rmse, 4),
        'Erro Absoluto Médio': round(mae, 4),
        'R2 Score': round(r2, 4)
    }

    return metricas
```

```python
calcular_metricas_regressao(y_test, y_pred_dummy)
```

## `Teinamento do *RandomForestREgressor*`

```python
#Aqui iremos importar um modelo mais avançado para que possamos treina-lo

from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor(max_depth=5, random_state=42)
model_rf.fit(X_train, y_train)
y_pred_rf = model_rf.predict(X_test)
calcular_metricas_regressao(y_test, y_pred_rf)
```

Nota-se uma melhora muito boa em relaçao ao modelo anterior, avaliado pela R2-score que avalia a confiabilidade do modelo. Alem disso se reduziu o erro absoluto medio e a raiz do erro quadratico medio.

```python
#Agora vamos importar uma classe de uma biblioteca, que fornece graficos
#para analise de modelos de regressao.O que vamos utilizar plota erros de predicao.

from yellowbrick.regressor import prediction_error

visualizer = prediction_error(model_rf, X_train, y_train, X_test, y_test)
```

Linha cinza era o ideal de prediçao. Linha preta foi o melhor fit

```python
#Agora vamos importar uma biblioteca que plota grafico de residuos(erro de predicao).
#O Residuals Plot vai fornecer justamente um gráfico de resíduos e um histograma mostrando essa distribuição dos resíduos.

#Residuo é o erro de previsão, o valor previsto menos o valor real

from yellowbrick.regressor import residuals_plot

viz = residuals_plot(model_rf, X_train, y_train, X_test, y_test)
```

Para o primeiro gráfico, à esquerda, os valores devem estar distribuídos aleatoriamente em torno da linha 0. Quanto mais distribuídos aleatoriamente em torno dessa linha, melhor é o modelo. Se os dados estiverem muito concentrados nessa linha, isso significa que o modelo deve ser repensado. No nosso caso, percebemos que está bem distribuído aleatoriamente: Temos pontos mais abaixo do gráfico, outros mais acima, uns mais à direita, uns mais à esquerda e assim por diante.

O gráfico que está à direita, é um histograma de resíduos. Já sabemos como interpretar um histograma, mas, para esse, o ideal é que ele esteja distribuído o mais uniformemente possível em torno do 0.

Se estiver muito assimétrico à esquerda e muito assimétrico à direita, isso significa que o nosso modelo não está legal. Isso porque o resíduo é a diferença entre o valor previsto e o valor real. Se estiver mais próximo do 0, significa que o nosso modelo está acertando mais, está tendo mais assertividade nas respostas, ou chegando o mais próximo possível da resposta. Então, como o nosso gráfico está, de certa forma, até bem próximo da distribuição normal, isso significa que o modelo está se adequando bem aos nossos dados.

##`Validaçao Cruzada`

```python
from sklearn.model_selection import KFold, cross_validate

scoring = {
    'mae': 'neg_mean_absolute_error',
    'rmse': 'neg_root_mean_squared_error',
    'r2': 'r2'
}

cv = KFold(n_splits=5, shuffle=True, random_state=42)

cv_results = cross_validate(model_rf, X_train, y_train, cv=cv, scoring=scoring)
cv_results
```

```python
for metric in scoring.keys():
    scores = cv_results[f'test_{metric}']
    mean_score = "{:.3f}".format(scores.mean())
    std_score = "{:.3f}".format(scores.std())

    print(f"{metric.upper()} Scores: {[f' {val:.3f}' for val in scores]}")
    print(f"{metric.upper()} Média: {mean_score}, Std: {std_score}")
    print('----------------------------------------------------------------------------------')
```

Dados muito proximos do que ja haviamo obtido

##`Otimizaçao de Hiperparamentros`

###`Seleçao de Recursos`

```python
#Imprime as features que estao tendo mais relevancia na construçao do modelo
from yellowbrick.model_selection import FeatureImportances

viz = FeatureImportances(model_rf, realtive=False, topn=10)
viz.fit(X_train, y_train)
viz.show()
```

```python
importances = model_rf.feature_importances_
feature_importances = pd.DataFrame({'Feature': X.columns, 'Importances': importances})
feature_importances = feature_importances.sort_values('Importances', ascending=False)
feature_importances
```

```python
results_df = pd.DataFrame(index=['RMSE', 'MAE', 'R2'])

model_selected_features = RandomForestRegressor(random_state=42, max_depth=5)

for count in [1, 5, 10, 15, 20, 25, 30]:
    selected_features = feature_importances['Feature'].values[:count]

    X_train_selected = X_train[selected_features]
    X_test_selected = X_test[selected_features]

    model_selected_features.fit(X_train_selected, y_train)

    y_pred = model_selected_features.predict(X_test_selected)

    metricas = calcular_metricas_regressao(y_test, y_pred)

    results_df[count] = list(metricas.values())

results_df
```

```python
results_df = pd.DataFrame(index=['RMSE', 'MAE', 'R2'])

model_selected_features = RandomForestRegressor(random_state=42, max_depth=5)

for count in range(10,16):
    selected_features = feature_importances['Feature'].values[:count]

    X_train_selected = X_train[selected_features]
    X_test_selected = X_test[selected_features]

    model_selected_features.fit(X_train_selected, y_train)

    y_pred = model_selected_features.predict(X_test_selected)

    metricas = calcular_metricas_regressao(y_test, y_pred)

    results_df[count] = list(metricas.values())

results_df
```

Nota-se que a partir da 13 feature os alores permanecem inalterados

## `Otimizando os hiperparametros com o GridSearchCV`

```python
param_grid = {
    'max_depth': [5, 10, 15],
    'min_samples_leaf': [1, 2, 3],
    'min_samples_split': [2, 4, 6],
    'n_estimators': [100, 150, 200]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)

model_grid = GridSearchCV(RandomForestRegressor(random_state=42), param_grid=param_grid, scoring='r2', cv=cv)

model_grid.fit(X_train, y_train)
```

```python
model_grid.best_params_
```

```python
y_pred_model_grid = model_grid.predict(X_test)

metricas_model_grid = calcular_metricas_regressao(y_test, y_pred_model_grid)
metricas_model_grid
```

```python
results_df['model_grid'] = list(metricas_model_grid.values())
results_df
```

Obtivemos um modelo melhor do que o anterior.
